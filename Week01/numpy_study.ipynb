```python
import numpy as np
np.random.seed(0)
arr = np.random.randint(0,20,size=(5,5))
print(arr)
print(arr[arr>10])
arr = np.where(arr>10, 0, arr)
print(arr)
```

```python
arr = np.array([120,135,110,125,130,140])
print(np.sum(arr))
print(np.mean(arr))
print(np.var(arr))
print(np.argmax(arr)+1)
print(np.argmin(arr)+1)
```

```python
A = np.array([1,2,3])
B = np.array([4,5,6])
C = np.concatenate((A,B))
print(C)
C = np.reshape(C, (2, 3))
print(C)
```
