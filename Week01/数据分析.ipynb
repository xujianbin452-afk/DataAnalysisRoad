```python
#缺失值处理
import pandas as pd
import numpy as np

df = pd.DataFrame({
    '语文': [90, np.nan, 70],
    '数学': [80, 60, 90],
    '英语': [np.nan, 75, 85]
}, index=['小王', '小李', '小张'])
print(df)
#print(df.dropna())
print(df.dropna())
print(df.dropna(axis=1))
print(df.dropna(how='all'))
print(df.dropna(thresh=3))
print(df.dropna(thresh=3, axis=1))
print(df.isna().sum(axis=1))
print(df.dropna(subset=['语文']))
```

```python
#填充缺失值
df = pd.read_csv('data/weather_withna.csv')
print(df.tail())
print(df.isna().sum())
#print(df.fillna({'temp_max':20, 'temp_min':10}).tail())
#print(df.fillna(df[['wind', 'precipitation']].mean()).tail())
print(df.fillna(df['wind'].mean()).tail())
print(df.ffill().tail())
```

```python
import pandas as pd
data = {
    'ID': [1, 2],
    'name':['alice','bob'],
    'Math': [90, 85],
    'English': [88, 92],
    'Science': [95, 89]
}
df = pd.DataFrame(data)
print(df)
df.T
df2 = df.melt(id_vars=['ID','name'], var_name='Subject', value_name='Score')
print(df2.sort_values('name'))
df3 = df2.pivot(index=['ID','name'], columns='Subject', values='Score').reset_index()
print(df3)
```
