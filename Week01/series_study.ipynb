```python
np.random.seed(0)
s = pd.Series(np.random.randint(50, 101, 10),index=['学生'+str(i) for i in range(1, 11)], name='成绩')
print(s)
print(s.mean())
print(s.max())
print(s.min())
print(s[s > s.mean()].count())
```

```python
temperature = pd.Series([28,31,29, 32, 30, 27, 33], index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday', 'Sunday'], name='Temperature')
print(temperature)
print(temperature[temperature > 30].count())
t = temperature.mean()
print(f'平均温度：{t}')
print(temperature.sort_values(ascending=False))
t2 = temperature.diff().abs().sort_values(ascending=False).index[2:3].tolist()
print(t2)
```

```python
date = pd.date_range(start='2023-01-01',periods=10)
prices = pd.Series([102.3,103.5,105.1,104.8,106.2,107.0,106.5,108.1,109.3,110.2], index=date, name='价格')
print(prices)
p2 = prices.pct_change().round(4)
print(p2)
print(f'收益率最高的日期：{p2.idxmax()},收益率最低的日期：{p2.idxmin()}')
print(f'波动率为:{p2.std():.4f}')
```

```python
sales = pd.Series([120,135,145,160,155,170,180,175,190,200,210,220], index=pd.date_range('2022-01-01', periods=12, freq='ME'), name='Sales')
#print(sales)
#print(sales.resample('QE').mean())
#print(f'销售额最高的月份：{sales.idxmax()}')
s2 = sales.pct_change().round(4)
#print(s2)
s3 = s2>0
print(s3)
s4 = s3[s3.rolling(window=3).sum()==3].keys().tolist()
print(f'连续三个月销售额增长的月份：{s4}')
print(s4)
```

```python
np.random.seed(1)
hourly_sales = pd.Series(np.random.randint(0,100,24),index = pd.date_range('2025-01-01',periods = 24,freq = 'h'))
#print(hourly_sales)
print(hourly_sales.resample('D').sum())
s1 = hourly_sales.values[8:23].sum()
print(f'8点到22点的销售额总和:{s1}')
s2 = hourly_sales.between_time('08:00', '22:00').sum()
print(f'8点到22点的销售额总和:{s2}')
#print(f'8点到22点的销售额总和:{s2.iloc[0]}')
print(s2/(hourly_sales.sum()-s2))
s3 = hourly_sales.sort_values(ascending = False)
print(f'销售额最高的三小时：{s3.index[0:3].tolist()}')

```
