```python
import pandas as pd
s = pd.Series([1, 2, 3, 4, 5])
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s)
```

```python
print(s['a'])
print(s[s > 3])
print(s.head())
```

```python
import numpy as np
s = pd.Series([1, 2, np.nan,3, None,4, 5], index=['a', 'b', 'c', 'd', 'e', 'f', 'g'],name='MySeries')
print(s)
```

```python
s.head()
```

```python
s.describe()
```

```python
print(s.isna())
print(s.index)
print(s.keys())
```

```python
print(s.isin([4, 5]))
print(s.sort_index())
print(s.sort_values())
```

```python
print(s.quantile(0.25))
print(s.quantile(0.5))
print(s.quantile(0.75))
print(s.sort_values(ascending=False))
```

```python
s['H'] = 4
print(s)
print(s.mode())
print(s.value_counts())
```

```python
s.drop_duplicates()
print(s)
s.drop_duplicates(inplace=True)
print(s)
s.unique()
print(s)
print(s.nunique())
```

```python
np.random.seed(0)
s = pd.Series(np.random.randint(50, 101, 10),index=['学生'+str(i) for i in range(1, 11)], name='成绩')
print(s)
print(s.mean())
print(s.max())
print(s.min())
print(s[s > s.mean()].count())
```
