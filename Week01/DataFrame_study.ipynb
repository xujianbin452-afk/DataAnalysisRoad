```python
data = {
    '数学':[85, 92, 78, 88, 95],
    '英语':[90, 88, 85, 92, 80],
    '物理':[75, 80, 88, 85, 90]
}
scores = pd.DataFrame(data,index=['张三','李四','王五','赵六','钱七'])
print(scores)
s2 = scores.T
print(s2)
#print(s2.sum(),s2.mean())
s3 = scores[(scores['数学']>90) | (scores['英语']>85)]
print(s3)
s4 = s2.sum()
print(s4.sort_values(ascending=False).head(3).index.tolist())
#print(s4.sort_index(s2.sum(), ascending=False))
```

```python
data = {
    '产品名称': ['A', 'B', 'C', 'D'],
    '单价': [100, 150, 200, 120],
    '销量': [50, 30, 20, 40]
}
df = pd.DataFrame(data)
print(df)
df['总销售额'] = df['单价'] * df['销量']
#df['累加'] = df[['单价', '销量', '总销售额']].sum(axis=1)
print(df)
df2 = df.nlargest(1,columns='总销售额')
print(df2)
print(f'总销售额最高的产品是：{df2["产品名称"].values[0]}，总销售额为：{df2["总销售额"].values[0]}')
print(df.sort_values(by='总销售额', ascending=False))
```
