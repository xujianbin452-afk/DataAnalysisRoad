```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family']= 'SimHei'  # 设置中文字体为SimHei

plt.figure(figsize=(10,5))


month = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
sales = [1000, 1200, 1500, 1300, 1700, 1600, 1800, 2000, 2200, 2100, 2300, 2500]
plt.plot(month, sales,label='产品A',color='orange',linewidth=2,linestyle='--',marker='o')
#添加标题
plt.title('2024年每月销售额',color='red',fontsize='20')
#添加坐标轴标签
plt.xlabel('月份',fontsize='10')
plt.ylabel('销售额（万元）',fontsize='10')
#添加图例
plt.legend(loc='upper left')
#添加网格线
plt.grid(True,alpha=0.3,linestyle='--')
#设置y轴范围
plt.ylim(1000,3000)
#x,y,刻度： plt.xticks()
#在每个数据点显示数据
for x,y in zip(month,sales):
    plt.text(x,y+1,str(y),ha='center',va='bottom')
#自动优化排版
plt.tight_layout()
#显示
plt.show()
```

```python
#温度分析
#1 导入库
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'
#2 导入数据
df = pd.read_csv('data/weather.csv')
print(df.head())
#df.info()
df['date'] = pd.to_datetime(df['date'])
#df.info()
df = df[df['date'].dt.year==2015]
#气温趋势图
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['temp_max'],label='最高气温')
plt.plot(df['date'],df['temp_min'],label='最低气温')
plt.title('2015气温趋势图')
plt.grid(True,alpha=0.7,linestyle='--')
plt.legend(loc='upper left',fontsize=20)
plt.xlabel('年月',fontsize = '10')
plt.ylabel('温度',fontsize = '10')
```
